```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

flame_data = pd.read_csv('/content/drive/MyDrive/Machine Learning data sets/flame.csv')
toy_data = pd.read_csv('/content/drive/MyDrive/Machine Learning data sets/toy.csv')

flame_array = flame_data.values
toy_array = toy_data.values

def kmeans(data, k, max_iterations=100):

    np.random.seed(0)
    centroids = data[np.random.choice(range(len(data)), size=k, replace=False)]

    for _ in range(max_iterations):
        distances = np.linalg.norm(data[:, np.newaxis] - centroids, axis=-1)
        labels = np.argmin(distances, axis=1)
        new_centroids = np.array([np.mean(data[labels == i], axis=0) for i in range(k)])

        if np.all(centroids == new_centroids):
            break

        centroids = new_centroids

    return labels, centroids

k = 2
flame_labels, flame_centroids = kmeans(flame_array, k)
toy_labels, toy_centroids = kmeans(toy_array, k)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(flame_array[:, 0], flame_array[:, 1], c='b', label='Cluster 1')
plt.scatter(flame_array[:, 2], flame_array[:, 2], c='r', label='Cluster 2')
plt.title('Original Clusters - Flame Dataset')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(toy_array[:, 0], toy_array[:, 1], c='b', label='Cluster 1')
plt.scatter(toy_array[:, 2], toy_array[:, 2], c='r', label='Cluster 2')
plt.title('Original Clusters - Toy Dataset')
plt.legend()

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(flame_array[:, 0], flame_array[:, 1], c=flame_labels)
plt.scatter(flame_centroids[:, 0], flame_centroids[:, 1], c='red', marker='x')
plt.title('Clusters Obtained by K-means - Flame Dataset')

plt.subplot(1, 2, 2)
plt.scatter(toy_array[:, 0], toy_array[:, 1], c=toy_labels)
plt.scatter(toy_centroids[:, 0], toy_centroids[:, 1], c='red', marker='x')
plt.title('Clusters Obtained by K-means - Toy Dataset')

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

flame_data = pd.read_csv('/content/drive/MyDrive/Machine Learning data sets/flame.csv')
toy_data = pd.read_csv('/content/drive/MyDrive/Machine Learning data sets/toy.csv')

flame_array = flame_data.values
toy_array = toy_data.values

k = 2
flame_kmeans = KMeans(n_clusters=k, random_state=0,init='random', max_iter=100,).fit(flame_array)
toy_kmeans = KMeans(n_clusters=k, random_state=0,init='random', max_iter=100,).fit(toy_array)

flame_labels = flame_kmeans.labels_
toy_labels = toy_kmeans.labels_

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(flame_array[:, 0], flame_array[:, 1], c='b', label='Cluster 1')
plt.scatter(flame_array[:, 2], flame_array[:, 2], c='r', label='Cluster 2')
plt.title('Original Clusters - Flame Dataset')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(toy_array[:, 0], toy_array[:, 1], c='b', label='Cluster 1')
plt.scatter(toy_array[:, 2], toy_array[:, 2], c='r', label='Cluster 2')
plt.title('Original Clusters - Toy Dataset')
plt.legend()

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(flame_array[:, 0], flame_array[:, 1], c=flame_labels)
plt.scatter(flame_kmeans.cluster_centers_[:, 0], flame_kmeans.cluster_centers_[:, 1], c='red', marker='x')
plt.title('Clusters Obtained by K-means - Flame Dataset')

plt.subplot(1, 2, 2)
plt.scatter(toy_array[:, 0], toy_array[:, 1], c=toy_labels)
plt.scatter(toy_kmeans.cluster_centers_[:, 0], toy_kmeans.cluster_centers_[:, 1], c='red', marker='x')
plt.title('Clusters Obtained by K-means - Toy Dataset')

plt.show()
```

```python
import pandas as pd
import warnings
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score, silhouette_score

# Load the dataset
data = pd.read_csv('/content/drive/MyDrive/Machine Learning data sets/shop.csv')

# Extract the features for clustering
features = data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

# Plot the data before clustering
plt.figure(figsize=(10, 6))
plt.scatter(features['Annual Income (k$)'], features['Spending Score (1-100)'], alpha=0.8)
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Data Before Clustering')
plt.show()

# Find the optimal number of clusters using the elbow method
inertia_values = []
silhouette_scores = []
max_clusters = 10  # Maximum number of clusters to consider

for k in range(2, max_clusters + 1):
    with warnings.catch_warnings():
        warnings.filterwarnings("ignore", category=FutureWarning)
        kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
        kmeans.fit(features)
    inertia_values.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(features, kmeans.labels_))

# Determine the optimal number of clusters based on the elbow method
elbow_index = -1
for i in range(1, len(inertia_values) - 1):
    if inertia_values[i] - inertia_values[i + 1] < 0.1 * (inertia_values[i - 1] - inertia_values[i]):
        elbow_index = i + 1
        break

optimal_num_clusters = elbow_index + 1 if elbow_index > 0 else 2

# Perform clustering with the optimal number of clusters
with warnings.catch_warnings():
    warnings.filterwarnings("ignore", category=FutureWarning)
    kmeans = KMeans(n_clusters=optimal_num_clusters, n_init=10, random_state=42)
    kmeans.fit(features)

# Evaluate clustering performance metrics
rand_index = adjusted_rand_score(data['cluster'], kmeans.labels_)
silhouette_index = silhouette_score(features, kmeans.labels_)

# Plot the data after clustering
plt.figure(figsize=(10, 6))
colors = ['blue', 'green', 'red', 'cyan', 'magenta', 'yellow', 'black', 'orange', 'pink', 'purple']
for i in range(optimal_num_clusters):
    cluster_points = features[kmeans.labels_ == i]
    plt.scatter(cluster_points['Annual Income (k$)'], cluster_points['Spending Score (1-100)'],
                color=colors[i], alpha=0.8, label=f'Cluster {i+1}')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='black', marker='x', s=150,
            label='Cluster Centers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Data After Clustering')
plt.legend()
plt.show()

# Print the results
print(f'Optimal number of clusters: {optimal_num_clusters}')
print(f'Rand Index: {rand_index:.4f}')
print(f'Silhouette Index: {silhouette_index:.4f}')
```
